```python
movie_db = catalog_df[['catalog_id', 'title', 'genres', 'director', 'actors', 'date_last_diff', 'start_rights', 'end_rights', 'original_language', 'valid_num_rights_TV', 'last_diff_rating', 'last_diff_rating_7', 'tmdb_id']]
```

```python
movie_db[movie_db['tmdb_id'].isna()]
```

```python
movie_db['tmdb_id'] = movie_db['tmdb_id'].astype('Int64')
```

```python
import numpy as np
type(np.nan)
```

```python
from tqdm.auto import tqdm
import api.tmdb 

tqdm.pandas()

movie_db['poster_path'] = movie_db['tmdb_id'].progress_apply(lambda mid: api.tmdb.get_movie_poster_path(mid, width=200))
```
